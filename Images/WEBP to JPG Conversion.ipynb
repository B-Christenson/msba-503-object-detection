```python
from PIL import Image
import os

folder = r"C:\Users\Brandon\Documents\MSBA\MSBA_in_progress_courses\MSBA_503_Analytics_programming_2\Take Home Assignment\Images"

for filename in os.listdir(folder):
    if filename.lower().endswith(".webp"):
        webp_path = os.path.join(folder, filename)
        jpg_path = os.path.join(folder, filename.rsplit(".", 1)[0] + ".jpg")

        img = Image.open(webp_path).convert("RGB")
        img.save(jpg_path, "JPEG", quality=95)
        print(f"Converted: {filename} → {os.path.basename(jpg_path)}")
```
